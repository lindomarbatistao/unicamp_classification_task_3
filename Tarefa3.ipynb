#**ENSEMBLE**

```python
# Primeiro, classifique os exemplos de teste usando cada um dos classificadores: árvore de decisão, Naïve Bayes e k-NN.
# Em seguida, atribua uma classe para cada exemplo baseada na maioria dos votos dos classificadores.
# A classe atribuída ao exemplo será a classe que obteve mais votos entre os classificadores.
# Repita este processo para todos os exemplos de teste.
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
x_treino, x_teste, y_treino, y_teste = train_test_split(previsores, dadosY2, test_size=0.3, random_state=60)
```

```python
# Treinar classificadores
tree_clf = DecisionTreeClassifier()
tree_clf.fit(x_treino, y_treino)
```

```python
nb_clf = GaussianNB()
nb_clf.fit(x_treino, y_treino)
```

```python
knn_clf = KNeighborsClassifier(n_neighbors=25, metric='minkowski', p=2)
knn_clf.fit(x_treino, y_treino)
```

```python
# Classifique os exemplos de teste usando cada classificador
tree_predictions = tree_clf.predict(x_teste)
nb_predictions = nb_clf.predict(x_teste)
knn_predictions = knn_clf.predict(x_teste)
```

```python
# Monte o ensemble com voto majoritário
ensemble_predictions = []
for i in range(len(x_teste)):
    votes = [tree_predictions[i], nb_predictions[i], knn_predictions[i]]
    # Calcula a classe com mais votos
    predicted_class = max(set(votes), key=votes.count)
    ensemble_predictions.append(predicted_class)
```

```python
# Avalie a acurácia do ensemble
ensemble_accuracy = accuracy_score(y_teste, ensemble_predictions)
print(f"Acurácia do ensemble com voto majoritário: {ensemble_accuracy*100:.2f}%")
```
